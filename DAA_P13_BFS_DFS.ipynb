```python
import networkx as nx
import matplotlib.pyplot as plt

hierarchy = {
    "Rektor": ["Warek 1", "Warek 2"],
    "Warek 1": [],
    "Warek 2": ["Kaprodi 1", "Kaprodi 2", "Kaprodi 3"],
    "Kaprodi 1": ["Dosen A", "Dosen B", "Dosen C"],
    "Kaprodi 2": ["Dosen D", "Dosen E"],
    "Kaprodi 3": ["Dosen F", "Dosen G"],
    "Dosen A": [],
    "Dosen B": [],
    "Dosen C": [],
    "Dosen D": [],
    "Dosen E": [],
    "Dosen F": [],
    "Dosen G": []
}

def dfs(graph, start, visited=None):
    if visited is None:
        visited = []
    visited.append(start)
    print(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

    return visited

def visualize_graph(graph):
    G = nx.DiGraph()

    for node, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)
    pos = {
        "Rektor": (0, 5),
        "Warek 1": (-1, 4), "Warek 2": (0, 4),
        "Kaprodi 1": (-1, 3), "Kaprodi 2": (0, 3), "Kaprodi 3": (1, 3),
        "Dosen A": (-2, 2), "Dosen B": (-1.5, 2), "Dosen C": (-1, 2),
        "Dosen D": (-0.5, 2), "Dosen E": (0.5, 2),
        "Dosen F": (1, 2), "Dosen G": (1.5, 2)
    }

    plt.figure(figsize=(12, 8))
    nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=3000, font_size=10, font_weight='bold', arrowsize=15)
    plt.title("Visualisasi Hierarki dengan DFS", fontsize=16)
    plt.show()
visited_nodes = dfs(hierarchy, "Rektor")
visualize_graph(hierarchy)
```

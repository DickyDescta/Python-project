```python
# Import library yang dibutuhkan
import networkx as nx

# Definisikan graph berdasarkan gambar
G = nx.Graph()

# Tambahkan simpul dan edge
nodes = [1, 2, 3, 4, 5, 6]
edges = [
    (1, 2), (1, 5),
    (2, 3), (2, 4),
    (3, 6), (4, 3),
    (5, 6)
]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Hitung centrality
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)

# Cetak hasil
print("Degree Centrality:")
print(degree_centrality)

print("\nBetweenness Centrality:")
print(betweenness_centrality)

print("\nCloseness Centrality:")
print(closeness_centrality)

print("\nEigenvector Centrality:")
print(eigenvector_centrality)
```

```python
# Import library yang dibutuhkan
import networkx as nx

# Definisikan graph berdasarkan gambar
G = nx.Graph()

# Tambahkan simpul dan edge
nodes = ["Mike", "Bob", "Emma", "Jill", "John", "Shane", "Leah", "Liz", "Allen", "Lisa"]
edges = [
    ("Mike", "Bob"), ("Mike", "Emma"), ("Mike", "Jill"),
    ("Bob", "Jill"), ("Bob", "John"),
    ("Emma", "Jill"), ("Emma", "Liz"),
    ("Jill", "John"), ("Jill", "Shane"), ("Jill", "Leah"),
    ("John", "Shane"), ("Shane", "Leah"),
    ("Liz", "Allen"), ("Allen", "Lisa")
]

G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Hitung centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

# Cetak hasil
print("Eigenvector Centrality:")
print(eigenvector_centrality)
```

```python
# 3. **Latihan: Membuat Network dan Menghitung Eigenvector Centrality**
# Penyelesaian latihan untuk membuat network kecil berdasarkan diagram yang diberikan

# Import library yang dibutuhkan
import networkx as nx
import matplotlib.pyplot as plt

# Membuat network kecil berdasarkan diagram yang diberikan
G_small = nx.Graph()
# Menambahkan edge sesuai diagram
edges = [
    (1, 2), (1, 4), (1, 10),
    (2, 3), (2, 11), (2, 15), (2, 12),
    (3, 5), (3, 13), (6, 7), (6, 8),
    (6, 9), (6, 14), (6, 2)
]
G_small.add_edges_from(edges)

# Visualisasi network kecil
plt.figure(figsize=(8, 8))
nx.draw_networkx(G_small, with_labels=True, node_color='orange', node_size=500)
plt.title("Network Kecil")
plt.show()

# Menghitung eigenvector centrality
eigencent = nx.eigenvector_centrality(G_small)
print("Eigenvector Centrality:")
for node, centrality in eigencent.items():
    print(f"Node {node}: {centrality:.4f}")
```

```python
# Import library yang dibutuhkan
import networkx as nx
import matplotlib.pyplot as plt
from google.colab import files

# Meminta pengguna untuk mengunggah file
print("Silakan unggah file facebook_combined.txt.gz Anda:")
uploaded = files.upload()

# Mendapatkan nama file yang diunggah
file_path = list(uploaded.keys())[0]
print(f"File {file_path} berhasil diunggah.")

# Membaca file data network
G_fb = nx.read_edgelist(file_path, create_using=nx.Graph(), nodetype=int)

# Menampilkan statistik dasar
print("Jumlah Node:", G_fb.number_of_nodes())
print("Jumlah Edge:", G_fb.number_of_edges())

# Membuat subgraph untuk percepatan analisis
# Mengambil subgraph dengan 500 node pertama
nodes_subset = list(G_fb.nodes())[:500]
G_small = G_fb.subgraph(nodes_subset).copy()

# Menampilkan statistik subgraph
print("Subgraph Jumlah Node:", G_small.number_of_nodes())
print("Subgraph Jumlah Edge:", G_small.number_of_edges())

# Visualisasi awal subgraph
plt.figure(figsize=(10, 10))
pos = nx.spring_layout(G_small, seed=42)  # Menggunakan spring layout untuk tata letak graf
nx.draw_networkx(G_small, pos, with_labels=False, node_size=10, edge_color="gray")
plt.title("Visualisasi Subgraph Facebook")
plt.show()

# Menghitung degree centrality untuk subgraph
degree_centrality = nx.degree_centrality(G_small)

# Menentukan node size berdasarkan degree centrality
node_size = [v * 1000 for v in degree_centrality.values()]

# Visualisasi dengan ukuran berdasarkan degree centrality
plt.figure(figsize=(10, 10))
nx.draw_networkx(
    G_small, pos, with_labels=False, node_size=node_size, node_color=list(degree_centrality.values()), cmap=plt.cm.plasma
)
plt.title("Visualisasi Berdasarkan Degree Centrality (Subgraph)")
plt.axis('off')
plt.show()
```

#Pseudocode
1. Mulai dari indeks pertama dari deret huruf.
2. Periksa setiap elemen satu per satu:
*  a. Jika elemen tersebut adalah "a", tampilkan pesan "Huruf 'a' ditemukan pada indeks ke-i".
*    b. Jika elemen tersebut bukan "a", lanjutkan ke elemen berikutnya.
3. Jika sudah memeriksa seluruh elemen dan "a" tidak ditemukan, tampilkan pesan "Huruf 'a' tidak ditemukan".
4. Selesai.

```python
#jawaban soal no 2
# Deret huruf
letters = ['y', 'u', 'i', 'w', 'o', 'a', 'q', 'u', 'j', 'p']

def linear_search(letters, target):
    for i in range(len(letters)):
        if letters[i] == target:
            return f"Huruf '{target}' ditemukan pada indeks ke-{i}"
    return f"Huruf '{target}' tidak ditemukan"
result = linear_search(letters, 'a')
print(result)

```

```python
#jawaban soal 3.2
def divide_and_conquer_min_max(arr, low, high):
    if low == high:
        return arr[low], arr[low]
    elif high == low + 1:
        if arr[low] < arr[high]:
            return arr[low], arr[high]
        else:
            return arr[high], arr[low]
    mid = (low + high) // 2
    min1, max1 = divide_and_conquer_min_max(arr, low, mid)
    min2, max2 = divide_and_conquer_min_max(arr, mid + 1, high)
    min_result = min(min1, min2)
    max_result = max(max1, max2)

    return min_result, max_result

A = [4, 12, 23, 9, 21, 1, 35, 2, 24]

minimum, maximum = divide_and_conquer_min_max(A, 0, len(A) - 1)
print("Nilai Minimum:", minimum)
print("Nilai Maximum:", maximum)
```

```python
#jawaban soal 4.b
import heapq
from collections import Counter
import matplotlib.pyplot as plt
import networkx as nx

class HuffmanNode:
    def __init__(self, char=None, freq=0, left=None, right=None):
        self.char = char
        self.freq = freq
        self.left = left
        self.right = right

    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(text):
    frequency = Counter(text)
    heap = [HuffmanNode(char, freq) for char, freq in frequency.items()]
    heapq.heapify(heap)

    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        merged = HuffmanNode(None, left.freq + right.freq, left, right)
        heapq.heappush(heap, merged)

    return heap[0]

def add_edges(graph, node, pos, x=0, y=0, layer=1):
    if node is not None:
        if node.char:
            label = f"{node.char}:{node.freq}"
        else:
            label = f"{node.freq}"

        graph.add_node(label, pos=(x, y))
        pos[label] = (x, y)

        if node.left is not None:
            left_x = x - 1 / layer
            left_y = y - 1
            graph.add_edge(label, f"{node.left.char}:{node.left.freq}" if node.left.char else f"{node.left.freq}")
            add_edges(graph, node.left, pos, x=left_x, y=left_y, layer=layer + 1)

        if node.right is not None:
            right_x = x + 1 / layer
            right_y = y - 1
            graph.add_edge(label, f"{node.right.char}:{node.right.freq}" if node.right.char else f"{node.right.freq}")
            add_edges(graph, node.right, pos, x=right_x, y=right_y, layer=layer + 1)

data = "BCAADDDCACAACAC"
root = build_huffman_tree(data)

G = nx.DiGraph()
pos = {}
add_edges(G, root, pos)

plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=10, font_weight="bold", font_color="black", edge_color="gray")
plt.title("Huffman Tree")
plt.show()
```

Menentukan Huffman Code untuk Setiap Huruf

*    A : 11
*    B : 100
*    C : 0
*    D : 101



```python
#jawaban soal 4.2
import heapq
from collections import Counter, namedtuple
class HuffmanNode:
    def __init__(self, char=None, freq=0, left=None, right=None):
        self.char = char
        self.freq = freq
        self.left = left
        self.right = right

    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(text):
    frequency = Counter(text)
    heap = [HuffmanNode(char, freq) for char, freq in frequency.items()]
    heapq.heapify(heap)

    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        merged = HuffmanNode(None, left.freq + right.freq, left, right)
        heapq.heappush(heap, merged)

    return heap[0]

def huffman_code_tree(node, prefix="", code={}):
    if node.char is not None:
        code[node.char] = prefix
    else:
        huffman_code_tree(node.left, prefix + "0", code)
        huffman_code_tree(node.right, prefix + "1", code)
    return code

data = "BCAADDDCACAACAC"

root = build_huffman_tree(data)
huffman_codes = huffman_code_tree(root)

print("Huffman Code untuk setiap huruf:")
for char, code in huffman_codes.items():
    print(f"{char}: {code}")
```
